## Import Data

```python
import pandas as pd
import numpy as np
```

```python
products= pd.read_csv("./products.csv")
products.head()
```

```python
ordered = pd.read_csv("./order_products__prior.csv")
ordered.head()
```

```python
ordered=ordered[['order_id','product_id']]
ordered.head()
```

```python
ordered
```

```python
products=products[['product_id','product_name']]
products.head()
```

```python
mba_data=ordered.merge(products, on="product_id", how="left")
```

```python
del mba_data['product_id']
```

```python
mba_data
```

```python
(mba_data.groupby('order_id').agg('count')).describe().transpose()
```

```python
g = mba_data.groupby('order_id')

a=np.arange(g.ngroups)
np.random.shuffle(a)

mba_data=mba_data[g.ngroup().isin(a[:1000000])]# change 2 to what you need :-) 
```

```python
mba_data=mba_data.sort_values('order_id')
```

```python
mba_data
```

```python
(mba_data.groupby('order_id').agg('count')).describe().transpose()
```

```python
mba_data.to_csv("./market_basket_train.csv", index = False)
```
